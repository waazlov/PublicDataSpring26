```python
import pandas as pd
import numpy as np
```

```python
# Load data
df = pd.read_csv("../data/raw/Traffic_Study_Flow_Counts.csv")

# Basic check
print(df.shape)
print(df.columns.tolist())
df.head()
```

```python
# Filter to 2026 only
df_2026 = df[df["STDY_YEAR"] == 2026].copy()

print("Rows in 2026:", len(df_2026))
```

```python
# Kep only columns needed for this assignment
df_2026 = df_2026[["O_STREET", "X_STREET", "STUDY_ADT"]].copy()

df_2026.head()
```

```python
# Clean street text
# Change value type to string and remove whitespaces just in case
df_2026["O_STREET"] = df_2026["O_STREET"].astype(str).str.strip()
df_2026["X_STREET"] = df_2026["X_STREET"].astype(str).str.strip()

# Turn fake "nan" strings back to real missing values so they can be removed easily
df_2026[["O_STREET", "X_STREET"]] = df_2026[["O_STREET", "X_STREET"]].replace("nan", np.nan)

# Clean traffic values
df_2026["STUDY_ADT"] = pd.to_numeric(df_2026["STUDY_ADT"], errors = "coerce")

# Remove rows missing needed values
df_2026 = df_2026.dropna(subset = ["O_STREET", "STUDY_ADT"]).copy()

print("Rows after cleaning: ", len(df_2026))
df_2026.head()
```

```python
# Create one location label for the chart
df_2026["LOCATION"] = np.where(
    df_2026["X_STREET"].notna(),
    df_2026["O_STREET"] + " & " + df_2026["X_STREET"],
    df_2026["O_STREET"]
)

df_2026.head()
```

```python
# Aggregate by location taking mean for same location entries having multiple ADT records
traffic_by_location = (
    df_2026.groupby("LOCATION", as_index = False)
    .agg(avg_adt = ("STUDY_ADT", "mean"))
)

traffic_by_location.head()
```

```python
# Sort from highest to lowest traffic
traffic_by_location = traffic_by_location.sort_values(
    by = "avg_adt",
    ascending = False
).reset_index(drop = True)

traffic_by_location["avg_adt"] = traffic_by_location["avg_adt"].round(2)

traffic_by_location.head(20)
```

```python
# Keep only top 15 for cleaner Flourish bar chart
top_15_traffic = traffic_by_location.head(15).copy()

top_15_traffic
```

```python
# Save aggregated and cleaned csv file for Flourish
top_15_traffic.to_csv("../data/processed/top_15_traffic_locations_2026.csv", index = False)
```
