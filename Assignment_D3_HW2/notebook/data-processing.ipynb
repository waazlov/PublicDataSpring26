```python
import pandas as pd
from pathlib import Path
```

```python
# Set file paths
raw_path = Path("../data/raw/Lyneham_Global_Summary_of_the_Year_Climate_data.csv")
processed_path = Path("../data/processed/lyneham_temperature_processed.csv")
```

```python
# Load the raw climate dataset
df = pd.read_csv(raw_path)
```

```python
# Keep only the columns needed for the multi-line chart
temperature = df[["DATE", "TAVG", "TMAX", "TMIN"]].copy()
```

```python
# Rename columns for easier use in D3
temperature = temperature.rename(columns={
    "DATE": "year",
    "TAVG": "tavg",
    "TMAX": "tmax",
    "TMIN": "tmin"
})
```

```python
# Convert year and temperature columns to numeric values
temperature["year"] = pd.to_numeric(temperature["year"])
temperature["tavg"] = pd.to_numeric(temperature["tavg"])
temperature["tmax"] = pd.to_numeric(temperature["tmax"])
temperature["tmin"] = pd.to_numeric(temperature["tmin"])
```

```python
# Remove rows with missing values in the chart variables
temperature = temperature.dropna(subset=["year", "tavg", "tmax", "tmin"])
```

```python
# Sort the data by year for the line chart
temperature = temperature.sort_values("year")
```

```python
# Convert year to integer after removing missing values
temperature["year"] = temperature["year"].astype(int)
```

```python
# Save the cleaned dataset as an external CSV for D3
temperature.to_csv(processed_path, index=False)
```
